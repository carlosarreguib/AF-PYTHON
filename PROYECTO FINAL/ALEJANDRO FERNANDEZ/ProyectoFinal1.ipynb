```python
class Libro:
    def __init__(self, titulo, autor, anoPublicacion, numeroPaginas):
        self.titulo = titulo
        self.autor = autor
        self.anoPublicacion = anoPublicacion
        self.numeroPaginas = numeroPaginas
        self.estado = 'disponible'

    def prestar(self):
        if self.estado == 'disponible':
            self.estado = 'prestado'
        else:
            print('ERROR: El libro no está disponible')
    
    def devolver(self):
        if self.estado == 'prestado':
            self.estado = 'disponible'
        else:
            print('ERROR: El libro no está prestado')
    
    def info(self):
        print('El libro {}, del autor {}, publicado en el año {}, con {} páginas, se encuentra {}'.format(self.titulo, self.autor, self.anoPublicacion, self.numeroPaginas, self.estado))

libros = []

def agregarLibro():
    titulo = input('Introduzca el título del libro: ')
    autor = input('Introduzca el autor del libro: ')
    anoPublicacion = int(input('Introduzca el año de publicación del libro: '))
    numeroPaginas = int(input('Introduzca el número de paginas del libro: '))
    libros.append(Libro(titulo, autor, anoPublicacion, numeroPaginas))
    print('¡Libro añadido correctamente!')

def consultarLibros():
    contador = 0
    if len(libros) > 0:
        print('Los libros disponibles son los siguientes:')
        for libro in libros:
            if libro.estado == 'disponible':
                print(libro.info())
                contador += 1
    else:
        print('Actualmente no hay ningún libro registrado')

def buscarLibros():
    contador = 0
    opcion1 = input('Introduzca cómo desea buscar los libros: titulo o autor ')
    opcion2 = input('Introduzca el {}: '.format(opcion1))
    match opcion1:
        case 'titulo':
            for libro in libros:
                if libro.titulo == opcion2:
                    print(libro.info())
                    contador += 1
        case 'autor':
            for libro in libros:
                if libro.autor == opcion2:
                    print(libro.info())
                    contador += 1
    if contador == 0:
        print('No se ha encontrado ningún resultado')

def guardarLibros():
    fichero = open('libros.txt', 'w')
    for libro in libros:
        fichero.write(libro.titulo + ';' + libro.autor + ';' + str(libro.anoPublicacion) + ';' + str(libro.numeroPaginas) + ';' + libro.estado + '\n')
    libros.clear()
    fichero.close()

def recuperarLibros():
    fichero = open('libros.txt')
    lineas = fichero.readlines()
    for linea in lineas:
        titulo, autor, anoPublicacion, numeroPaginas, _ = linea.split(';')
        libros.append(Libro(titulo, autor, int(anoPublicacion), int(numeroPaginas)))
    fichero.close()

def totalPaginas():
    contador = 0
    for libro in libros:
        contador += libro.numeroPaginas
    print('El número total de páginas es {}'.format(contador))

def promedioPaginas():
    contador = 0
    for libro in libros:
        contador += libro.numeroPaginas
    print('El número promedio de paginas por libro es {}'.format(contador/len(libros)))

def librosEnAno():
    ano = int(input('Introduzca un año: '))
    contador = 0
    for libro in libros:
        if ano == libro.anoPublicacion:
            contador += 1
    if contador > 0:
        print('{} libros han sido publicados en el año {}'.format(contador, ano))
    else:
        print('No se han encontrado libros publicados en el año {}'.format(ano))


def mostrarMenu():
    while 1:
        opcion = input('Introduzca lo que desea hacer: AGREGAR LIBRO (agregar) CONSULTAR LIBROS DISPONIBLES (consultar) BUSCAR LIBROS (buscar) GUARDAR LIBROS (guardar) RECUPERAR LIBROS (recuperar) PAGINAS TOTALES (paginas) PROMEDIO PAGINAS (promedio) LIBROS EN AÑO (ano) SALIR (salir): ')
        match opcion:
            case 'agregar':
                agregarLibro()
            case 'consultar':
                consultarLibros()
            case 'buscar':
                buscarLibros()
            case 'guardar':
                guardarLibros()
            case 'recuperar':
                recuperarLibros()
            case 'paginas':
                totalPaginas()
            case 'promedio':
                promedioPaginas()
            case 'ano':
                librosEnAno()
            case 'salir':
                break

mostrarMenu()
```
