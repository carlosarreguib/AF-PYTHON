```python
# Crea un programa que gestione la información de una biblioteca.
# Para ello, tendrás que implementar una clase "Libro" que tenga los
# siguientes atributos: título, autor, año de publicación, número de
# páginas y estado (disponible o prestado). Además, esta clase deberá
# contar con métodos para prestar y devolver el libro, así como para
# imprimir su información.
# El programa deberá permitir al usuario agregar nuevos libros a la
# biblioteca, buscar libros por título o autor, y listar todos los libros
# disponibles. También deberá contar con la funcionalidad de guardar
# la información de la biblioteca en un archivo de texto y cargarla desde
# dicho archivo.
# Además, deberás implementar las siguientes funciones utilizando
# programación funcional:
# - Una función que calcule la cantidad total de páginas en la
# biblioteca.
# - Una función que calcule el promedio de páginas por libro.
# - Una función que calcule la cantidad de libros publicados en un
# año determinado.
# Por último, deberás incluir la gestión de excepciones para evitar
# errores en la entrada de datos por parte del usuario, así como para
# manejar posibles errores al leer o escribir el archivo de la biblioteca.



class Libro:
    def __init__(self, titulo, autor, fecha, paginas, estado):
        self.titulo = titulo
        self.autor = autor
        self.fecha = fecha
        self.paginas = paginas
        self.estado = estado

    def prestar():
        estado = "prestado"
    def devolver():
        estado = "disponible"

def leer_fichero():
  db = open("db.txt", "r")
  datos = db.read()
  lista = datos.split("\n")
  db.close()
  return(lista)

def menu_biblioteca(db):
  ans=True
  while ans:
      print("""
      1. Nuevo libro
      2. Consultar
      3. Listar todos los libros
      4. Reservar libro
      5. Devolver libro
      6. Cargar base de datos db.txt
      7. Guardar información en db.txt
      8. Estadisticas de la biblioteca
      9. Salir
      """)
      ans=input("Selecciona una opción ")
      if ans=="1":
        nuevo(db)
      elif ans=="2":
        consultar(db)
      elif ans=="3":
        listar(db)
      elif ans=="4":
        reservar(db)
      elif ans=="5":
        devolver(db)
      elif ans=="6":
        db = cargar(db)
      elif ans=="7":
        guardar(db)
      elif ans=="8":
        estadisticas(db)
      elif ans=="9":
        print("Hasta pronto.\n") 
        ans = None
      else:
        print("\n Opción inválida!")

def nuevo(db):
    print('Añadir nuevo libro al repertorio.')
    titulo = input("Introduce el título: ")
    autor = input("Introduce el autor: ")
    fecha = input("Introduce la fecha: ")
    paginas = input("Introduce el número de páginas: ")
    try:
      book = Libro(titulo, autor, int(fecha), int(paginas), "disponible")
      db.append(book.titulo+";"+book.autor+";"+str(book.fecha)+";"+str(book.paginas)+";"+book.estado)
    except ValueError:
      print("Introduce unicamente numeros en la fecha y las paginas")

def consultar(db):
    print('Has elegido la opción 2: Consulta de libros')
    tipobusqueda = input('Elige un método de busqueda\n 1. Por autor\n 2. Por título\n')
    if (tipobusqueda == "1"):
      autores = list(map(lambda x: x.split(';')[1], db))
      respuesta = input("Escribe el autor: ")
      try:
        indice = autores.index(respuesta)
        print(db[indice])
      except ValueError:
        print("No se ha encontrado el valor introducido en la base de datos.")
    elif (tipobusqueda == "2"):
      titulos = list(map(lambda x: x.split(';')[0], db))
      respuesta = input("Escribe el titulo: ")
      try:
        indice = titulos.index(respuesta)
        print(db[indice])
      except ValueError:
        print("No se ha encontrado el valor introducido en la base de datos.")
    else:
      print("Selección inválida")

def listar(db):
    print('Has elegido la opción 3: Listar todos los libros')
    try:
      libros = list(map(lambda x: x.split(';')[0], db))
      print('Catálogo de libros:\n', *libros, sep='\n- ')
    except TypeError:
      print('No hay libros en la base de datos. Introducelos o carga la BD desde un fichero.')

def reservar(db):
    print('Has elegido la opción 4: Reserva de libros.')
    titulos = list(map(lambda x: x.split(';')[0], db))
    respuesta = input("Escribe el titulo: ")
    try:
      indice = titulos.index(respuesta)
      ejemplar = db[indice].split(";")
      disp = ejemplar[4]
      if (disp == "disponible"):
        disp = "reservado"
        db[indice] = ejemplar[0]+";"+ejemplar[1]+";"+ejemplar[2]+";"+ejemplar[3]+";"+disp
        print("Libro reservado correctamente")
      else:
        print("El libro no está disponible.")
    except ValueError:
      print("No se ha encontrado el valor introducido en la base de datos.")
    return db
def devolver(db):
    titulos = list(map(lambda x: x.split(';')[0], db))
    respuesta = input("Escribe el titulo: ")
    try:
      indice = titulos.index(respuesta)
      ejemplar = db[indice].split(";")
      disp = ejemplar[4]
      if (disp == "reservado"):
        disp = "disponible"
        db[indice] = ejemplar[0]+";"+ejemplar[1]+";"+ejemplar[2]+";"+ejemplar[3]+";"+disp
        print("Libro devuelto correctamente")
      else:
        print("El libro seleccionado no estaba reservado.")
    except ValueError:
      print("No se ha encontrado el valor introducido en la base de datos.")
def cargar(db):
    print('Has elegido la opción 6: Cargar base de datos')
    confirmacion = input('Esta opción sustituye los datos en memoria por los del fichero db.txt. ¿Quieres continuar con el proceso? (S/N): ')
    if (confirmacion == "S"):
      db = leer_fichero()
      print('Se ha cargado el fichero db.txt en el programa.')
    return db
    
def guardar(db):
    print('Has elegido la opción 7: Guardar información')
    fich = open('db.txt','w')
    for libro in db:
      fich.write(libro+"\n")
    fich.close()

def estadisticas(db):
  # - Una función que calcule la cantidad total de páginas en la
  # biblioteca.
  paginas = list(map(lambda x: x.split(';')[3], db))
  years = list(map(lambda x: x.split(';')[2], db))
  numpaginas = 0
  numlibros = 0
  for ejemplar in paginas:
    numlibros = numlibros + 1
    numpaginas = numpaginas + int(ejemplar)
  print("El número total de páginas de la biblioteca es de " + str(numpaginas))
  # - Una función que calcule el promedio de páginas por libro.
  print("La media de páginas de la biblioteca es de " + str(numpaginas/numlibros))
  # - Una función que calcule la cantidad de libros publicados en un
  # año determinado.
  print("La cantidad de libros publicada en 2020 es de " + years.count("2021"))
if (db):
  db = menu_biblioteca(db)
else:
  db = []
  menu_biblioteca(db)
```
